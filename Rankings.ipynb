```python
import pandas as pd
```

```python
df = pd.read_csv('rankingsch.csv',encoding = 'latin1')
```

```python
df.sample(5)
```

```python
lst1 = list(df.DeathRate)
lst2 = list(df.State)
ranks = []
for i in range(len(lst1)):
    lst1[i] = lst1[i].strip()

if len(lst1) == len(lst2):    
    for i in range(len(lst1)):
        rank = lst1.index(lst2[i])+1
        ranks.append(rank)
    df['DeathRateRanking'] =  ranks
else:
    print("ERROR: Number of states not same in the columns")
```

### Adding Premium:

```python
df1 = pd.read_csv('WithPremium.csv')
df1.sample(5)
```

```python
states = list(df1.State)
for i in range(len(states)):
    states[i] = states[i].strip()
premium = list(df1.Average_Premium)
state_dict = dict(zip(states,premium))
```

```python
states = list(df.State)
by_state_premium = [0]*len(states)
for i in range(len(states)):
    by_state_premium[i] = state_dict[states[i]]
```

```python
df['AveragePremium'] = by_state_premium
```

```python
df.drop(['DeathRate','HealthSpending','Income','InsuredPeople'],axis=1,inplace=True)
df.to_csv("WithRankings.csv")
```
